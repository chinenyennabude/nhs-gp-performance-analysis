```python
file_path = "../data/raw/GP_Appointment_December_2025.xlsx"
sheet = "Table 1a"

# read a small preview without headers
preview = pd.read_excel(file_path, sheet_name=sheet, header=None, nrows=30)
preview.head(15)
```

```python
# Find the row index that contains a month label like "Dec-25"
header_row = None
for i in range(len(preview)):
    row_as_text = preview.iloc[i].astype(str).tolist()
    if any("Dec-25" in x for x in row_as_text):  # you can change to "Nov-25" etc.
        header_row = i
        break

header_row
```

```python
df = pd.read_excel(
    file_path,
    sheet_name=sheet,
    header=header_row
)

df.head()
```

```python
import pandas as pd

file_path = "../data/raw/GP_Appointment_December_2025.xlsx"
sheet = "Table 1a"

df = pd.read_excel(file_path, sheet_name=sheet, header=10)
df.head()
```

```python
from datetime import datetime

new_cols = []
for c in df.columns:
    # If column name is a datetime/Timestamp, format it
    if isinstance(c, (datetime, pd.Timestamp)):
        new_cols.append(pd.to_datetime(c).strftime("%b-%y"))  # Dec-25
    else:
        new_cols.append(c)

df.columns = new_cols
df.head()
```

```python
# first two columns are label columns in this table
section_col = df.columns[0]
item_col = df.columns[1]

df = df.dropna(how="all")
df[section_col] = df[section_col].ffill()
df = df[df[item_col].notna()]  # drop separator rows

df.head(30)
```

```python
month_cols = df.columns[2:]  # all the month columns

long_df = df.melt(
    id_vars=[section_col, item_col],
    value_vars=month_cols,
    var_name="Month",
    value_name="Value"
)

long_df.head(10)
```
