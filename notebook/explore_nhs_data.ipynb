```python
import pandas as pd
from datetime import datetime, timedelta
import re

# Created this function which encapsulates the logic to read, clean, and reshape the NHS GP appointments data
def clean_gp_table_to_long(
    file_path: str,
    sheet_name: str,
    n_preview_rows: int = 80,
    min_month_hits: int = 4,   # how many month-like cells to treat a row as the header
) -> pd.DataFrame:
    """
    Reads an NHS GP appointments Excel sheet (e.g., 'Table 1a'), auto-detects the header row
    by finding the row with multiple month/date-like headers, cleans it, and returns a
    long-format dataframe.
    """

    # 1) Preview to detect header row
    preview = pd.read_excel(file_path, sheet_name=sheet_name, header=None, nrows=n_preview_rows)

    month_pat = re.compile(r"^[A-Za-z]{3}-\d{2}$")          # Jul-23
    month_in_text_pat = re.compile(r"([A-Za-z]{3}-\d{2})")  # Jul-23, 7, 8

    def excel_serial_to_date(n):
        return datetime(1899, 12, 30) + timedelta(days=int(n))

    def is_month_like(x) -> bool:
        # True if x looks like a month header in any common NHS format
        if pd.isna(x):
            return False

        # datetime / timestamp
        if isinstance(x, (datetime, pd.Timestamp)):
            return True

        # Excel serial date (usually large, but allow > 0 defensively)
        if isinstance(x, (int, float)):
            # Filter out tiny numbers that are likely footnotes unless they convert cleanly
            if x <= 0:
                return False
            try:
                _ = excel_serial_to_date(x)
                return True
            except Exception:
                return False

        # strings: allow Jul-23 or Jul-23, 7, 8 etc
        s = str(x).strip()
        if month_pat.match(s):
            return True
        if month_in_text_pat.search(s):
            return True

        return False

    # Score each row by how many "month-like" cells it contains
    best_row = None
    best_score = 0

    for i in range(len(preview)):
        score = sum(is_month_like(v) for v in preview.iloc[i].tolist())
        if score > best_score:
            best_score = score
            best_row = i

    if best_row is None or best_score < min_month_hits:
        raise ValueError(
            f"Could not confidently detect a header row in '{sheet_name}'. "
            f"Best score={best_score}. Try increasing n_preview_rows or lowering min_month_hits."
        )

    header_row = best_row

    # 2) Read the sheet using detected header row
    df = pd.read_excel(file_path, sheet_name=sheet_name, header=header_row)

    # 3) Standardise column headers into clean 'MMM-YY' where applicable
    def standardise_month_header(col):
        # datetime header
        if isinstance(col, (datetime, pd.Timestamp)):
            return pd.to_datetime(col).strftime("%b-%y")

        # Excel serial header
        if isinstance(col, (int, float)) and col > 0:
            try:
                return excel_serial_to_date(col).strftime("%b-%y")
            except Exception:
                return str(col)

        # string header with potential footnotes
        s = str(col).strip()
        m = month_in_text_pat.search(s)
        if m:
            return m.group(1)

        return s

    df.columns = [standardise_month_header(c) for c in df.columns]

    # 4) Identify label columns
    section_col = df.columns[0]
    metric_col = df.columns[1]

    # Drop empty rows and clean label columns
    df = df.dropna(how="all")
    df[section_col] = df[section_col].ffill()
    df = df[df[metric_col].notna()]

    # 5) Melt wide → long
    month_cols = df.columns[2:]
    long_df = df.melt(
        id_vars=[section_col, metric_col],
        value_vars=month_cols,
        var_name="month",
        value_name="value"
    ).rename(columns={section_col: "section", metric_col: "metric"})

    # 6) Keep only valid month labels (removes Unnamed columns)
    long_df["month"] = long_df["month"].astype(str).str.extract(r"([A-Za-z]{3}-\d{2})", expand=False)
    long_df = long_df.dropna(subset=["month"])

    # 7) Create month_date + numeric value
    long_df["month_date"] = pd.to_datetime(long_df["month"], format="%b-%y", errors="coerce")
    long_df["value"] = pd.to_numeric(long_df["value"], errors="coerce")

    return long_df
```
